```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
BB = pd.read_csv('BigBasket Products.csv')
print(BB.head())
```

```python
# Fill with mean (numeric)
BB['sale_price'] = BB['sale_price'].fillna(BB['sale_price'].mean())
BB['market_price'] = BB['market_price'].fillna(BB['market_price'].mean())
# Fill with mode (categorical)
BB['product'] = BB['product'].fillna(BB['product'].mode()[0])

# Drop rows with missing values
BB = BB.dropna()
```

```python
BB = BB.drop_duplicates()
```

```python
# Remove extreme values using percentile
q_low = BB['sale_price'].quantile(0.01)
q_high = BB['sale_price'].quantile(0.99)

df = BB[(BB['sale_price'] >= q_low) & (BB['sale_price'] <= q_high)]
```

```python
# Remove extreme values using percentile
q_low = BB['market_price'].quantile(0.01)
q_high = BB['market_price'].quantile(0.99)

df = BB[(BB['market_price'] >= q_low) & (BB['market_price'] <= q_high)]
```

```python
# Remove extra spaces
df['product'] = df['product'].str.strip()

# Convert to lowercase
df['product'] = df['product'].str.lower()
```

```python
# Example: Discount %
df['discount_percent'] = ((df['market_price'] - df['sale_price']) / df['market_price']) * 100
```

```python
df.to_csv("cleaned_bigbasket_data.csv", index=False)
```
